```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# 1. Load the csv file and show top 5 records from it.
```

```python
df = pd.read_csv("players_clean.csv")
df
# Created a dataframe for the csv file of soccer players and called it 'df'.
# To save time, I have limited the dataset to 1000 records out of 18000 and cleaned the file externally using powerBI and Excel.
# Cleaned: Null Values, unreadable text in 'short_name'/'long_name',blanks and unwanted info.
```

```python
# Top 5 records of players using '.head()' function.
top_five = df.head(5)
top_five
```

```python
# 2. How you would be able to see each column's name.
```

```python
# Called the 'sorted()' function on the dataframe which returned a list of the column names in the dataset.
sorted(df)
```

```python
# 3. Need to show number of rows and columns of this dataset.
```

```python
# Using '.shape' method on the dataframe returns the number of rows and columns in the dataset
# These results show that there are 1000 ROWS and 31 COLUMNS.
df.shape
```

```python
# 4. Show number of players and their countries.
```

```python
# Grouped players by their nationality by using 'groupby()', on'nationality' and counting the number of players by name
# Declared in 'countries'
#Results show that there are 91 countries that have FIFA soccer players

countries =df.groupby(['nationality'], as_index=False)['short_name'].count()
countries
```

```python
# 5. If you find many records in point 4 then show only top 10 countries and their number of players.
```

```python
# Using '.head()' on 'countries', shows only 10 nationalities of players out of 91.
# Declared in 'country_bar' to answer question 6.
country_bar = countries.head(10)
country_bar
```

```python
# 6. Now you have to create a bar plot of top 5 countries and their number of players, try to fill green color in bars.
```

```python
# Created bar graph showing the number of players from each country using 'country_bar'
# Changed bar color to green using 'color=green' in 'plt.bar'

plt.figure(figsize=(15,6))
plt.bar(country_bar['nationality'],country_bar['short_name'],color='green')
plt.xlabel('Nationality', size = 15)
plt.ylabel('Number of Players', size=15)
plt.title('Players per Country')
plt.show()
```

```python
# 7. Show top 5 players short name and wages.
```

```python
# Calling 'top_five' df from question 1, short_name and wage are filtered and declared in 'wage'
# Results show the first five players in the dataset and their wage

wage = top_five[['short_name','wage_eur']]
wage
```

```python
# 8. Show top 5 players short name and wages that are getting highest salaries.
```

```python
# filtered short names and wages and called 'nlargest()' function on wage to get 5 highest wages
# Declared in 'top_wage'

top_wage = df[['short_name','wage_eur']].nlargest(5,'wage_eur')
top_wage
```

```python
# 9. Create a bar plot of point number 8.
```

```python
# Bar chart shows the first 5 players with the highest wages from number 8 in purple using 'top_wage'

plt.figure(figsize=(10,6))
plt.bar(top_wage['short_name'],top_wage['wage_eur'],color='purple')
plt.xlabel('Players', size = 15)
plt.ylabel('Wage', size=15)
plt.title('Top 5 Highest Wages')
plt.show()
```
